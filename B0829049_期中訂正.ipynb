```python
from thinkdsp import CosSignal, SinSignal

cos_sig = CosSignal(freq=10, amp=10.0, offset=0)
sin_sig = SinSignal(freq=20, amp=10.0, offset=0)
```

```python
##(1.1)

from thinkdsp import decorate

cos_sig.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
##(1.2)

sin_sig.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix
```

```python
##(1.3)

mix.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
##(1.4)

wave = mix.make_wave(duration=2.00, start=0, framerate=11025)
wave
```

```python
wave.plot()
pl.grid()
```

```python
##(1.5) 
##訂正

wave2 = mix.make_wave(duration=2.00, start=0, framerate=1000)
wave2
```

```python
wave2.plot()
pl.grid()
```

```python
ans= wave2.ys.size
print(f'答案= {ans}')
```

```python
##(1.6)

spectrum = wave2.make_spectrum()
spectrum.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.7)
##訂正

x1= CosSignal(freq= 1000)
x2= CosSignal(freq= 2000)
x3= CosSignal(freq= 3000)
x4= CosSignal(freq= 4000)
y= x1+x2+x3+x4
wave3= y.make_wave()
spectrum= wave3.make_spectrum()
spectrum.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.8.1)
##訂正

spectrum1= spectrum.copy()
spectrum1.high_pass(2500)
spectrum1.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.8.2)
##訂正

spectrum2= spectrum.copy()
spectrum2.low_pass(2500)
spectrum2.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(3.1)

# Aliasing(混疊)是信號經取樣後頻譜發生重疊之現象，即高於採樣頻率一半的頻率成分將被重建成低於採樣頻率一半的訊號。而重建出來的訊號稱為原訊號的混疊替身，因為這兩個訊號有同樣的樣本值。
# 例如在音樂上訊號中會含有超出人耳聽力範圍的高頻成分，如果音樂訊號取樣低於32000取樣/秒(赫茲)任何等於16000赫茲或以上的訊號將在數位轉類比轉換器中造成混疊現象。
```
