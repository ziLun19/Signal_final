```python
import matplotlib.pyplot as pl
pl.grid(True)
```

```python
from thinkdsp import CosSignal, SinSignal

cos_sig = CosSignal(freq=10, amp=10.0, offset=0)
sin_sig = SinSignal(freq=20, amp=10.0, offset=0)
```

```python
##(1.1)

from thinkdsp import decorate

cos_sig.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
##(1.2)

sin_sig.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix
```

```python
##(1.3)

mix.plot()
pl.grid()
decorate(xlabel='Time (s)')
```

```python
##(1.4)

wave = mix.make_wave(duration=2.00, start=0, framerate=11025)
wave
```

```python
wave.plot()
pl.grid()
```

```python
##(1.5) 
##

wave2 = mix.make_wave(duration=2.00, start=0, framerate=1000)
wave2
```

```python
wave2.plot()
pl.grid()
```

```python
ans= wave2.ys.size
print(f'= {ans}')
```

```python
##(1.6)

spectrum = wave2.make_spectrum()
spectrum.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.7)
##

x1= CosSignal(freq= 1000)
x2= CosSignal(freq= 2000)
x3= CosSignal(freq= 3000)
x4= CosSignal(freq= 4000)
y= x1+x2+x3+x4
wave3= y.make_wave()
spectrum= wave3.make_spectrum()
spectrum.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.8.1)
##

spectrum1= spectrum.copy()
spectrum1.high_pass(2500)
spectrum1.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```

```python
##(1.8.2)
##

spectrum2= spectrum.copy()
spectrum2.low_pass(2500)
spectrum2.plot()
pl.grid()
decorate(xlabel='Frequency (Hz)')
```
